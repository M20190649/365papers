```python
%matplotlib notebook
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import glob
import datetime
import seaborn as sns
```

```python
files = glob.glob("../_posts/*.md")
dates = [
    datetime.datetime.strptime(
        "-".join(f.split("/")[-1].split('.')[0].split("-")[:-1]),
        "%Y-%m-%d"
    )
    for f in files
]
posts = [open(filename, 'r').read() for filename in files]
```

```python
contents = ["\n".join(p.split('---')[2:]).strip() for p in posts]
```

```python
plt.figure()
plt.hist2d([d.timetuple().tm_wday for d in dates], [len(c) for c in contents], bins=(7, 20))
plt.ylabel("Content Length")
plt.xlabel("Day of the week (0=sun)")
plt.title("Content Length by Weekday")
plt.xticks(range(7), ['sun', 'mon', 'tue', 'wed', 'thu', 'fri', 'sat'])
plt.show()
```

```python
plt.figure()
plt.title("Length of Content over Time")
plt.scatter([d for d in dates], [len(c) for c in contents])
```

```python
df = pd.DataFrame({"dates":dates, "contents":contents})
```

```python
plt.figure()
plt.title("Content Length Hist")
plt.hist(df.contents.str.len(), bins=40)
plt.show()
```

```python
print("Words per Entry\n\nmin:  {}\nmax:  {}\nmean: {}\nsum:  {}".format(
    min(df.contents.str.split().map(lambda x: len(x))), 
    max(df.contents.str.split().map(lambda x: len(x))), 
    int(np.mean(df.contents.str.split().map(lambda x: len(x)))), 
    sum(df.contents.str.split().map(lambda x: len(x)))
))
```
